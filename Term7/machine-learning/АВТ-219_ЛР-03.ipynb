````python
def PLOT34(y_test, y_pred) -> None:
    '''
    Функция построения графиков
    :param y_test: - проверочные значения целевой переменной
    :param y_pred: - вычисленные значения целевой переменной
    '''
    plt.figure(figsize=(12,6))

    # == Диаграмма рассеяния вычисленных значений ==
    # создать первое полотно 121: 1-строк, 2-столбцов, 1-индекс текущего полотна в сетке
    plt.subplot(121)
    plt.scatter(y_test, y_pred,  alpha=0.1, color = "#17becf")
    plt.axline((0, 0), slope=1, color='black', linestyle='--', linewidth=3, alpha=0.7,)
    range_test, range_pred = np.max(y_test)- np.min(y_test) , np.max(y_pred)- np.min(y_pred)
    if range_test/range_pred <4 and range_test/range_pred > 0.25:
        plt.gca().set_aspect('equal')
        axmin, axmax = np.min([y_test, y_pred]), np.max([y_test, y_pred])
        plt.xlim([axmin, axmax]); plt.ylim([axmin, axmax]);
    plt.title('Диаграмма рассеяния вычисленных значений');
    plt.xlabel('Проверочное Y');    plt.ylabel('Вычисленное Y');
    plt.grid(True,  alpha=0.5)   # Сетка

    # == Диаграмма рассеяния ошибок ==
    # создать второе полотно 121: 1-строк, 2-столбцов, 2-индекс текущего полотна в сетке
    plt.subplot(122)
    plt.scatter(y_test, (y_test - y_pred)**2,  alpha=0.1, color = "#17becf")
    plt.title('Диаграмма рассеяния квадрата абсолютной ошибки')
    plt.xlabel('Проверочное Y');    plt.ylabel('Квадрат абсолютной ошибки');
    plt.grid(True,  alpha=0.5)  # Сетка
#--------------------------------------------------------------------------
````

````python
def GET_METRICS_SINGLE(y_test, y_pred) -> None:
    '''
    Вычисление и вывод метрик: MAE, RMSE, R2. Используются функции из библиотеки sklearn. На основе сравнения проверочных и вычисленных значений.
    :param y_test: - проверочные значения целевой переменной
    :param y_pred: - вычисленные значения целевой переменной
    Возвращает словарь (хэш-таблицу) со значениями метрик
    '''
    mae  = metrics.mean_absolute_error        (y_test, y_pred)
    mse  = metrics.mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2   = metrics.r2_score                   (y_test, y_pred)
    print ("MAE : {:>9,.3f} (средняя абсолютная ошибка)".format              ( mae ))
    print ("MSE : {:>9,.6f} (среднеквадратичная ошибка)".format( mse ))
    print ("RMSE: {:>9,.6f} (кв. корень из среднеквадратичной ошибки)".format( rmse ))
    print ("R2  : {:>9,.3f} (коэфф. детерминации)".format                    ( r2 ))
    return {"MAE":mae, "MSE":mse, "RMSE":rmse, "R2":r2}
#--------------------------------------------------------------------------
````

Создадим переменную `X` как набор значений области определения функции в диапазоне  [-1.0, 1.0].

````python
N = 1000 # кол-во точек X
X = np.linspace(-1., 1., N).reshape(-1, 1) # привести массив к столбчатой структуре
````

Получим набор значения для имитации случайных отклонений для зависимой величины `y`.

````python
noiseLevel= 0.05 # степень зашумленности
Y_noise = np.random.normal(-noiseLevel,noiseLevel,N).reshape(-1, 1)
````

Объявим вспомогательные переменные для отображения дальнейших графиков.

````python
transparencyVal = 0.3  # Прозрачность точек графика
markerSize = 10        # Размер точки графика
````



Используя функции различного вида, получим значения `y` для каждого `x`.

````python
# Экспонента
Y_exp = np.exp(X).reshape(-1, 1) # привести массив к столбчатой структуре
Y_exp= Y_exp+Y_noise
plt.scatter(X, Y_exp, s=markerSize, color="black", alpha=transparencyVal)
````

````python
# Линейный с граничными значениями
c1 = int(X.size * 0.25)
c2 = int(X.size * 0.50)
c3 = X.size - c1 - c2

Y_linstep = np.concatenate([
        np.zeros( c1 ),
        np.linspace(start=0., stop=1., num=c2),
        1.0 + np.zeros(c3 )
        ],
    ).reshape(-1, 1)# привести массив к столбчатой структуре

Y_linstep= Y_linstep+Y_noise/2
plt.scatter(X, Y_linstep, s=markerSize, color="black", alpha=transparencyVal)
````

````python
# Полином
Y_polyn = (0.3*(X**3) + 0.1*(X**2) + 0.1 * X + 0.3).reshape(-1, 1) # привести массив к столбчатой структуре
Y_polyn = Y_polyn + Y_noise/2
plt.scatter(X, Y_polyn, s=markerSize, color="black", alpha=transparencyVal)
````

## Структура и параметры

Основным элементом искусственной нейронной сети (ИНС) является формальный нейрон. Выход нейрона формирует итоговый сигнал "y" в зависимости от входных сигналов x0,…,xN. Значения входных сигналов могут усиливаться или ослабляться в зависимости от величины и знака весовых (передаточных) коэффициентов синапсов w0,…,wN.
Взвешенные сигналы суммируются и могут поступать на вход некоторой функции преобразователя - функции активации (передаточной функции). Функции активации могут иметь различный вид: линейный, либо разнообразный нелинейный.


изображение.png

Для решения задачи восстановления числовой зависимости (регрессии) будем использовать ИНС со структурой многослойный перцептрон (MLP, Multi Layer Perceptron). В такой структуре сигналы распространяются от входов к выходам и нейроны сгруппированы в слои. При этом соседние слои являются полносвязными, т.е. каждый нейрон из слоя-источника передает сигналы в каждый нейрон слоя-приемника.

Используя таблицу с образцами (определяющие Х и зависимый от них У), есть возможность сравнить вычисленные значения y_pred и ожидаемые (проверочные) y_test. Ошибки (разности) между вычисленными и ожидаемыми значениями передают в функцию оценки потерь (loss function) - функцию тем или иным способом вычисляющую единый коэфф. несоответствия.

Используя текущие значения потерь (loss), выбранная функция оптимизации  вычислит поправки к текущим весовым коэфф. w0,…,wN и применит их. Поправки вычисляются таким образом, чтобы минимизировать значения потерь (loss).

В целом при подходе "обучение с учителем" нейронная сеть итеративно подбирает комбинацию весовых коэфф. такую, чтобы вычисленные ответы были как можно точнее в сравнении с обучающими образцами.

изображение.png

Нейронная сеть для решения задачи регрессии должна содержать:
- Входной слой `Input` с размером, равным количеству передаваемых аргументов Х;
- Один или более скрытый полносвязный слой `Dense`
- Выходной слой `Dense`, число нейронов которого равно числу желаемых ответов У. Выходной слой должен передавать ответ без изменений, т.е. не иметь функции активации

Различные типы слоёв в пакете `tensorflow.keras.layers` имеют множество параметров для настройки. В данной работе будут использованы следующие:

- Input: входной слой, не изменяет проходящие значения. Содержит параметр
  - shape=(x) - определяющий размерность входной матрицы
- Dense: полносвязный слой, получает информацию со всех узлов предыдущего слоя. Содержит параметры:
  - units - кол-во узлов;
  - activation - выбор встроенной функции активации ;
- Activation: обособленный слой с выбранной функцией активации.

Полный список типов слоёв смотрите на странице справки
https://www.tensorflow.org/api_docs/python/tf/keras/layers

Обратите внимание, что допустимо задавать функции активации (ФА) несколькими способами:
1. "Явный", при котором ФА определена в самостоятельном слое и распологается сразу после полносвязного слоя `Dense`. Пример:
    ```
    model.add( layers.Dense     (  units= 1, activation= None ) )
    model.add( layers.Activation(  activation= tf.keras.activations.relu ) )
    ```
2. "Компактный", при котором желаема ФА передается в виде параметра `activation` объекта `Dense`
    ```
    model.add(  layers.Dense    (  units= 1,  activation= tf.keras.activations.relu ) )
    ```
Код в приведённых выше примерах приводит к одному и тому же итоговому состоянию сети.

* Структура нейронной сети:
  - кол-во входных нейронов = кол-ву "x"
  - кол-во скрытых полносвязных слоев
  - кол-во нейронов на каждом скрытом слое
  - типы функций активации после каждого слоя
  - кол-во выходных нейронов = кол-ву "y"

* Функции активации для скрытых слоев в пакете `tensorflow.keras.activation`. Некоторые доступные функции активации нейронов:
    - None - отсутсвие функции активации, означает передачу сигнала без изменений
    - relu
    - sigmoid:  sigmoid(x) = 1 / (1 + exp(-x))
    - tanh
    - linear
    - elu
    - Полный список функций активации смотрите на странице справке https://www.tensorflow.org/api_docs/python/tf/keras/activations



* Функция оценки потерь (способ оценки ошибок между y_pred и y), она же носит название Loss function Для задачи регрессии подходит оценка Среднеквадратичной ошибки (MSE)

    Доступные функции оценки потерь смотрите на странице справки https://www.tensorflow.org/api_docs/python/tf/keras/losses


* Функция оптимизации потерь/ошибок, т.е. поиска коэффициентов, при которых потери/ошибки будут минимальны. В качестве такой функции вы можете выбрать `Adam` или `SGD` (Стохастический градиентный спуск)

    Доступные функции оптимизации смотрите на странице справки https://www.tensorflow.org/api_docs/python/tf/keras/optimizers

Для создания нейросети в библиотеке Tensorflow необходимо создать объект tensorflow.keras.models.Sequential, который будет содержать последовательность отдельных слоёв (layers). Этот класс содержит множество методов, некоторые из них:
- add(layer) - добавить слой
- compile(loss, optimizer) - проверить и подготовить к обучению, задать функцию оценки потерь и функцию оптимизации потерь
- summary - сводная инф-ия
- fit(X, Y, epoch) - обучить
- predict (X) - вычислить ответы
- save - сохранить модель в файл
- load_weights/save_weights

Полный список функции и их параметров см. на странице справки https://www.tensorflow.org/api_docs/python/tf/keras/Sequential

Учитывая выбранные выше параметры, составим код для создания НС.

Заполните пропущенные параметры, при необходимости добавьте дополнительные слои. При этом рекомендуется:
- Использовать как минимум один скрытый слой с количеством узлов > 30 шт. и нелинейной функцией активации

Обратите внимание на то, что для решения задачи регрессии выходной слой не должен иметь нелинейной функции активации.

````python
totalHistoryLossTrain=[] # Вспомогательный список для хранения полной истории обучения
totalHistoryLossTest=[]  # Вспомогательный список для хранения полной истории обучения
globalEpochCounter = 1   # Счётчик выполненных эпох обучения НС

###########################################
#▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼

# 2.1) Построение модели нейронной сети многослойного(Sequential) типа ----------------------------------------
# Многослойная
model = tf.keras.models.Sequential()

# ДОБАВЛЕНИЕ НОВЫХ СЛОЁВ --------------------------

# Входной слой
model.add( tf.keras.layers.Input(shape=(  1 ,))) # Входной слой

# Скрытые слои
#   параметр "units" - кол-во узлов/нейронов на данном слое
#   параметр "activation" - вид функции активации
model.add(tf.keras.layers.Dense(units= 1000 ,  activation=None,        )) # Скрытый слой, полносвязный, без встроенной нелинейной функции активации
model.add(tf.keras.layers.Activation( activation=tf.keras.activations.sigmoid ,        )) # Обособленный слой с выбранной нелинейной функцией активации

# Дополнительные скрытые слои
#...

# Выходной полносвязный слой, Без нелинейной функции активации
# Кол-во нейронов на последнем слое опредеяет кол-во выходных значений нейронной сети
model.add(tf.keras.layers.Dense(units=1,  activation=None))


# 2.2) Функции потерь и функции оптимизации --------------------------------------------------------------------
fLoss     =tf.keras.losses.MeanSquaredError()
#fOptimizer=tf.keras.optimizers.SGD(learning_rate=0.01)
fOptimizer=tf.keras.optimizers.Adam(learning_rate=0.01)
#fMetricList  = [tf.keras.losses.MeanSquaredError(), tf.keras.losses.MeanAbsoluteError()] # Дополнительные метрики для оценки ошибок/потерь

model.compile(
    loss      =fLoss, # Выбор функции оценки ошибок/потерь
    optimizer =fOptimizer, # Выбор функции для минимзации значения выбранной оценки ошибок/потерь
    #metrics   = fMetricList # Дополнительные метрики для оценки ошибок/потерь
)

print(model.summary())
#▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲
###########################################
````

## Первичное и дополнительное обучение

В процессе обучения будем получать вычисленные значения целевой переменной и строить график функции, соответствующий  текущему состоянию нейронной сети. Примеры изменения графиков представлены ниже.

Далее следует выполнить несколько циклов обучения (подстройки коэфф. каждого узла-нейрона) модели. Осмотрите результаты обучения, при необходимости выполните код нижестоящей ячейки повторно, при этом произойдёт дообучение нейросети. На графике также обратите внимание на численное значение выбранной функции оценки потерь (в примере используется fLoss =tf.keras.losses.mse).

_\*Для наглядности демонстрации процесса обучения и для возможности построения графика функции соответствующей текущему состоянию нейросети, количество эпох вынесено в цикл FOR, а соответствующий параметр epochs в метода fit установлен в "1"._


````python
# Выполнить заданное кол-во эпох обучения
epochForTrain = 300 # Кол-во эпох обучения
for currEpochNum in range(1, epochForTrain+1):
    # 3) Цикл обучения/продолжения обучения сети

    # Вызов метода для выполнения обучения (настройки коэфф. нейросети)
    history = model.fit(
        xNorm_train,  # Обучающие X
        yNorm_train,  # Обучающие Y
        #validation_data=(xNorm_test, yNorm_test), # Опционально проверочные X и Y, для вычисления оценок потерь (loss)
        epochs=10,       # Кол-во эпох обучения
        batch_size=100, # Внутри каждой эпохи, проводить обучение пакетами/порциями по batch_size строк
        verbose=0,
    )

    #  Каждые N эпох выводить график текущей функции НС. Только в условиях ЕДИНСТВЕННОГО x.
    if xNorm_train.shape[1] == 1:
        if (currEpochNum) % 10 == 0:
            yNorm_pred = model.predict(xNorm_train, verbose=0) # Вычислить ответы НС
            clear_output(wait=True) # Очистить окно вывода
            print(f'currEpochNum: {currEpochNum}/{epochForTrain}')
            plt.figure(figsize=(6, 4))
            plt.scatter(xNorm_train, yNorm_train,  s=markerSize, color='black', alpha=transparencyVal, label =  'Исходные данные')
            plt.scatter(xNorm_train, yNorm_pred,   s=markerSize, color='red',   alpha=transparencyVal, label = f'График функции нейронной сети. MSE {history.history["loss"][-1]:>9.6f}')
            plt.title(f'Выбранный нормализованный набор. ({globalEpochCounter:>4} эпоха)')
            plt.ylabel('Целевой Y'); plt.xlabel('Объясняющий X');
            plt.ylim( (-0.1, 1.1))
            plt.legend(loc='lower right')
            plt.grid(True, alpha=0.3)
            plt.show()
            plt.pause(0.05) # Временная пауза для отображения графика
    globalEpochCounter+=1

    # Дополнение полной истории обучения
    totalHistoryLossTrain.extend(history.history['loss'])
    if 'val_loss' in history.history.keys():
        totalHistoryLossTest.extend(history.history['val_loss'])
````

 В процессе обучения, НС на каждой эпохе измеряет и фиксирует значения ошибок/потерь (loss) с помощью выбранной функции. После выполнения метод `fit` возвращает объект `history` в котором можно получить доступ списку значениям этих ошибок, но только по тем эпохам, которые были выполнены в последнем вызове метода `fit`. В вышестоящей ячейке значения из `history` переносится в `totalHistoryLossTrain` и `totalHistoryLossTest` для получения полной истории с самой первой эпохи.

## Использование класса MinMaxScaler

В библиотеке sklearn, для выполнения процедуры нормализации, присутствует более удобный класс __sklearn.preprocessing.MinMaxScaler__, который содержит методы:

- __MinMaxScaler.fit(  df  )__ - вычисление значений __минимального__, __максимального значений__ и __диапазона__

- __MinMaxScaler.transform( df )__ - прямого преобразования из истинного значения к приведенному

- __MinMaxScaler.inverse_transform( df )__ - обратного преобразования из приведенного значения в истинное

## Структура и параметры

Дополните код для построения НС со структурой `2вх->1вых`, следовательно:
- входной слой должен принимать 2 значения
- скрытые слои отсутствуют
- выходной слой вычисляет единственное значение и не имеет функцию активации

````python
totalHistoryLossTrain=[] # Вспомогательный список для хранения полной истории обучения
totalHistoryLossTest=[]  # Вспомогательный список для хранения полной истории обучения

###########################################
#▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼
# 1) Параметры структуры

# 2.1) Построение модели нейронной сети многослойного(Sequential) типа ----------------------------
# Многослойная
m1 = tf.keras.models.Sequential()

# ДОБАВЛЕНИЕ НОВЫХ СЛОЁВ --------------------------

# Входной слой
m1.add(   tf.keras.layers.Input(
    shape=(  2 ,  ) # пример: shape=(1,)
  ))

# Скрытые слои
#   параметр "units" - кол-во узлов/нейронов на данном слое
#   параметр "activation" - вид функции активации
# Скрытый слой, полносвязный, без встроенной нелинейной функции активации
m1.add(   tf.keras.layers.Dense(    units     = 100    ,    activation=None,))
#  Обособленный слой с выбранной нелинейной функцией активации
m1.add(    tf.keras.layers.Activation(  activation = tf.keras.activations.relu, ))

# Дополнительные скрытые слои
#...

# Выходной полносвязный слой, Без нелинейной функции активации
# Кол-во нейронов на последнем слое опредеяет кол-во выходных значений нейронной сети
m1.add(    tf.keras.layers.Dense(units=1,  activation=None))



# 2.2) Функции потерь и оптимизации
fLoss     =tf.keras.losses.MeanSquaredError()

fOptimizer=tf.keras.optimizers.Adam(learning_rate=0.01) # SGD(learning_rate=0.01) или  Adam(learning_rate=0.01)
#fMetricList  =[tf.keras.losses.MeanSquaredError(), tf.keras.losses.MeanAbsoluteError()] # Дополнительные метрики для оценки ошибок/потерь

m1.compile(
    loss      =fLoss, # Выбор функции оценки ошибок/потерь
    optimizer =fOptimizer, # Выбор функции для минимзации значения выбранной оценки ошибок/потерь
    #metrics   = fMetricList # Дополнительные метрики для оценки ошибок/потерь
)

print(m1.summary())
#▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲
###########################################
````

````python
totalHistoryLossTrain=[] # Вспомогательный список для хранения полной истории обучения
totalHistoryLossTest=[]  # Вспомогательный список для хранения полной истории обучения

###########################################
#▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼▼
# 1) Параметры структуры

# 2.1) Построение модели нейронной сети многослойного(Sequential) типа ----------------------------
# Многослойная
m4 = tf.keras.models.Sequential()

# ДОБАВЛЕНИЕ НОВЫХ СЛОЁВ --------------------------

# Входной слой
m4.add(   tf.keras.layers.Input(
    shape=(  3 ,  ) # пример: shape=(1,)
  ))

# Скрытые слои
#   параметр "units" - кол-во узлов/нейронов на данном слое
#   параметр "activation" - вид функции активации
# Скрытый слой, полносвязный, без встроенной нелинейной функции активации
m4.add(   tf.keras.layers.Dense(    units     = 300    ,    activation=None,))
#  Обособленный слой с выбранной нелинейной функцией активации
m4.add(    tf.keras.layers.Activation(  activation = tf.keras.activations.relu, ))

# Дополнительные скрытые слои
#...
m4.add(   tf.keras.layers.Dense(    units     = 300    ,    activation=None,))
m4.add(   tf.keras.layers.Dense(    units     = 300    ,    activation=None,))

# Выходной полносвязный слой, Без нелинейной функции активации
# Кол-во нейронов на последнем слое опредеяет кол-во выходных значений нейронной сети
m4.add(    tf.keras.layers.Dense(units=1,  activation=None))



# 2.2) Функции потерь и оптимизации
fLoss     =tf.keras.losses.MeanSquaredError()

fOptimizer=tf.keras.optimizers.Adam(learning_rate=0.01) # SGD(learning_rate=0.01) или  Adam(learning_rate=0.01)
#fMetricList  =[tf.keras.losses.MeanSquaredError(), tf.keras.losses.MeanAbsoluteError()] # Дополнительные метрики для оценки ошибок/потерь

m4.compile(
    loss      =fLoss, # Выбор функции оценки ошибок/потерь
    optimizer =fOptimizer, # Выбор функции для минимзации значения выбранной оценки ошибок/потерь
    #metrics   = fMetricList # Дополнительные метрики для оценки ошибок/потерь
)

print(m4.summary())
#▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲▲
###########################################
````

````python
# 3) Обучение/продолжение обучения сети
epochForTrain = 300

# Вызов метода для выполнения обучения (настройки коэфф. нейросети)
history = m4.fit(
    xNorm_train,  # обучающие X  # требуется передать таблицу с кол-ом столбцов равным кол-ву входных нейронов
    yNorm_train,  # обучающие Y
    validation_data=(  # опционально: проверочные X и Y
        xNorm_test,  # требуется передать таблицу с кол-вом столбцов равным кол-ву входных нейронов
        yNorm_test),
    epochs=epochForTrain,   # кол-во эпох обучения
    #batch_size=100, # внутри каждой эпохи, проводить обучение пакетами/порциями по batch_size строк
    verbose=1,
)

# Дополнение полной истории обучения
totalHistoryLossTrain.extend(history.history['loss'])
if 'val_loss' in history.history.keys():
    totalHistoryLossTest.extend(history.history['val_loss'])
````

````python
plt.figure(figsize=(12,4))
plt.plot(totalHistoryLossTrain, label='train', color = '#1f77b4')
if 'val_loss' in history.history.keys():
    plt.plot(totalHistoryLossTest, label='test', color = '#ff7f0e')
plt.title("История изменений выбраной оценки потерь (LossVal = mean_squared_error)")
plt.ylabel("Оценка потерь (mean_squared_error)"); plt.xlabel("Эпохи обучения")
plt.legend()
plt.grid(True, alpha=0.5)  # Сетка
###plt.ylim ( (0, 0.03) ) # Область видимости для оси "Оценки потерь"
````
