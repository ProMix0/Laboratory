Для понимания как распределены отсутствующие значения в исходной таблице, попробуем получить графическое изображение __таблицы-маски__

```python
from matplotlib.colors import ListedColormap
def plotTableNA(dataF, addToTilte =""):
    '''  Функция для построения графика отображение отсутсвующих значений'''
    if   len(dataF.columns)> 20:   K = 3
    elif len(dataF.columns)== 1:   K = 1
    else:                          K = 1.7

    figSizeH = 15        # размер полотна, высота
    figSizeW =int(len(dataF.columns)/K) # размер полотна, ширина
    fig, ax = plt.subplots( figsize=(figSizeW, figSizeH)) # создать полотно для рисования, figsize-размер в дюймах

    # отрисовать матрицу значений функцией imshow. Применяется цветовая палитра
    plt.imshow( dataF.isna(),
                cmap = ListedColormap([ '#3B5A92', 'white', ]), # выбор цветовой шкалы, аналог cm.get_cmap('jet'),
                aspect='auto',  # ‘auto’ | ‘equal’ | scalar]    # режим соотнощения сторон
                interpolation= 'none',)                         # отключить размытие
    ax.set_xticks(np.arange(len(dataF.columns)))
    ax.set_xticklabels(dataF.columns, rotation=-70,)
    plt.title("Графическое отображение отсутсвующих значений (белые). "+ addToTilte)
    plt.grid(False) # Сетка
#-------------------------------------
# Вызов функции с указанием параметров
plotTableNA( dataF=df, addToTilte="Исходный" )
```

### One-Hot encoding

Используем приём известный как One-Hot encoding: cформируем новую таблицу (DataFrame), в которой будут добавлены столбцы по количеству уникальных значений категориального признака.

При этом только в каждой строке новой группы столбцов-индикаторов будет присутствовать  одно значение "1" как признак принадлежности к исходной категории, остальные значения равны "0".

изображение.png

```python
objColNames    = ['region'] # названия признаков в исходной таблицы в виде списка
prefixColNames = ['region']       # сокращенные названия признаков для новой таблицы в виде списка

df_dummies = pd.get_dummies(
    data   = df[objColNames],  # таблица с признаками для кодирования
    prefix = prefixColNames,   # сокращенные приставки к новым столбцам
    dtype  = int,              # результат сравнения в виде целых чисел {0,1}
)

df_dummies[:3] # осмотрим перые строки новой таблицы
```

## Корреляция

* __Коэффициент корреляции r-Пирсона__
  
  Мера линейной взаимосвязи переменных, при нелинейной взаимосвязи непоказателен. Не устойчив к выбросам в выборке. Учитывает числовые значения. Переменные должны иметь нормальное распределение. Изменяется в пределах от -1 до +1.

![Drag Racing](https://upload.wikimedia.org/wikipedia/commons/thumb/0/02/Correlation_examples.png/400px-Correlation_examples.png)

Построим матрицу корреляции - матрицу значений коэфф. корреляции Пирсона для возможных комбинай пар.

```python
corr_df = df_digits.corr(method='pearson')
corr_df
```

Для того, чтобы визуально оценить величину коэффициентов корреляции, постоим тепловую карту по значениям взаимозависимости признаков выборки.

```python
# построить "маску" для отображения только половины корреляционной матрицы
mask = np.zeros_like(corr_df, dtype=np.bool_)
mask[np.triu_indices_from(mask)] = True

# Построить тепловую карту
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(
            ##################
            data=corr_df,   ### <<<- объект-матрица корреляции
            ##################
            vmin=-1, vmax=1, center=0, # Диапазон цветовой шкалы
            cmap = 'RdBu_r', # Вариант цветовой шкалы
            annot=True, # Подписи в ячейках отобразить
            fmt='.2f', # Округление значений
            annot_kws={"size": 9}, # Размер шрифта подписей в ячейках
            mask=mask,   # Применить маску для отрисовки половины матрицы
            square=True, # Форма ячейки - квадрат
            linewidths=1.0, # Зазор между ячейки
            cbar_kws={"shrink": .7},   # Уменьшить размер легенды (цветовой шкалы)
           )
ax.set_xticklabels(ax.get_xticklabels(), rotation=-80) # Подписи горизонтальной оси развернуть
plt.title(u'Матрица корреляции признаков');
```

## Поиск и обработка выбросов

Используем график диаграммы размаха (ящик с усами, box-plot) для оценки наличия нетипичных наблюдений (выбросов, outliers) - значений, которые в наборе данных сильно удалены от медианного значения. На графике присутствуют следующие описательные статистики:
- Медиана (Q2, второй квартиль) — значение, которое разделяет упорядоченную выборку по количеству значений пополам;
- Q1 (первый квартиль) — медиана меньшей половины данных, т.е. 25% данных меньше этого значения;
- Q3 (третий квартиль) — медиана большей половины данных, т.е. 25% данных больше этого значения;
- IQR (межквартильный размах) — рассчитывается как интервал IQR = Q3 - Q1, в котором сосредоточена центральная 50% данных;
- Нетипичные наблюдения (выбросы, outliers) - значения, которые в наборе данных сильно удалены от медианного значения;
- Нижняя граница (ус) — рассчитывается как Q1 - 1.5*IQR
- Верхняя граница (ус) — рассчитывается как Q3 + 1.5*IQR


изображение.png

Судя по виду построенным диаграммам размаха (boxplot) нетипичные значения присуствуют. Для их фильтрации выберем граничные значения соответсвующие тонким краям ("усам").

Установим границы значений, для тех объектов таблицы, которые должны остаться

```python
# создадим "словарь"/"хэш-таблицу" для удобного обращения, где ключ - это название признака
tresholdDict = {
    'bmi'     : (0,44)            , # кол-во комнат
    'charges' : (0, 34000)     , # год постройки
}

print (tresholdDict)
```

Отфильтруем значения которые находятся вне пороговых.

```python
# Во избежание путаницы создадим для полседующих операций копию исходного объетка
df_digits_filtered = df_digits.copy()
for feature_name  in tresholdDict:
    if feature_name not in tresholdDict:
      continue
    treshMin =  tresholdDict[feature_name][0]
    treshMax =  tresholdDict[feature_name][1]
    # фильтр по двум границам. Оператор "~" -инверсия логических значений
    df_digits_filtered = df_digits_filtered[
        (df_digits_filtered[feature_name] >= treshMin ) & (df_digits_filtered[feature_name] <= treshMax)
    ]
    print("Обработан", feature_name )
print ("Отброшено {} строк. Осталось {} строк.".format(    len(df_digits)  - len(df_digits_filtered), len(df_digits_filtered)))
```

```python
n_plots = len(features+target) # Кол-во графиков
n_cols = 4 # Кол-во столбцов
n_rows = n_plots//n_cols + (n_plots % n_cols > 0) # Кол-во строк
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols*3.5, n_rows*3)) # Создаем полотно figure и axes
if n_rows == 1 and n_cols == 1:  axes = np.array([[axes]]) # Если axes не массив, преобразуем
elif n_rows == 1: axes = axes.reshape(1, -1)
elif n_cols == 1: axes = axes.reshape(-1, 1)

# Строим гистограммы по каждому признаку
for i, feature_name in enumerate(features+target):
    row, col= i // n_cols , i % n_cols # Номер строки, столбца графика
    ax = axes[row, col]
    sns.histplot(data=df_digits_filtered[feature_name],ax=ax) # Построение гистрограммы
    ax.set_title(feature_name) # Заголовок
    ax.grid(True, alpha=0.5) # Сетка
    ax.set_xlabel(u'Значения')
    ax.set_ylabel(u'Кол-во')
    print (feature_name,
           df_digits_filtered[feature_name].min(),
           df_digits_filtered[feature_name].max())
# Скрываем пустые subplot'ы
for i in range(n_plots, n_rows * n_cols):
    row, col= i // n_cols , i % n_cols # Номер строки, столбца графика
    axes[row, col].set_visible(False)

plt.tight_layout()
```

## Формирование тренировочной и проверочной выборок

Для обучения и проверки качества модели применяют прием заключающийся в разделении исходной выборки на части:

  - Обучающую (Train) - данные, на которых модель формирует коэффициенты.
  - Проверочную (Test) - данные, которые предъявляют обученной модели для получения вычисленных(Predicted) значений и последующего определения ошибок, по сравнению с истиными (ожидаемыми) ответами.

Для разделения воспользуемся функцией `train_test_split` из пакета sklearn. Принцип разделения показан на рис. ниже.


изображение.png

```python
valid_size = 0.3 # доля тестовой части в выборке
rand_seed = 42 # начальное состояние генератора случ. чисел

x_train, x_test, y_train, y_test =  train_test_split(
    df_digits_filtered[features],    # исходные данные X
    df_digits_filtered[target],      # исходные данные y

    test_size = valid_size, # доля тестовой части в выборке
    random_state=rand_seed, # начальное состояние генератора случ. чисел
    shuffle=True            # перемешивание
)
```

## Модель M1. Одномерная - от единственного признака 'bmi'

Для получения линейной регрессионной модели созданим экземпляр класса
__sklearn.linear_model.LinearRegression()__

```python
model_bmi = linear_model.LinearRegression()
```

Полученная модель требует процедуры обучения - подбора значений коэфф. для минимизации отличий между вычисленными значениями и предоставленными целевыми (обучающими y). У большинства моделей доступен метод __.fit(x_train, y_train)__

```python
# Вычислить коэфф. Ki в функции y(x)=Ki*xi + .... + B
# на тренировочном наборе, т.е. обучить модель
model_bmi.fit(
    x_train[ ['bmi', ] ], # x_train[['x1', 'x2', 'x3', ...]]
    y_train[ ['charges'] ]
)
```

Важным этапом является получение ответов от обученной модели, вычисленных на той части данных, которая не учавствовала в обучении (т.е не учавствовала в подстройке коэфф. мат. модели)

```python
# Получить вычисленные(predicted) зн. на проверочном наборе
# y_pred = fittedModel.predict(x_test)
y_predicted_on_bmi = model_bmi.predict(x_test[ ['bmi', ] ]) # x_train[['x1', 'x2', 'x3', ...]]
```

Для наглядности построим график рассеяния (он же график разброса, точечный, scatter) зависимой величины y от **единственного** определяющего признака

```python
# Создать полотно для рисования с указаными размерами
plt.figure(figsize=(15,5))

# Точечная диаграмма (диаграмма рассеяния) - plt.scatter
plt.scatter(x_train[['bmi']],  y_train[['charges']], alpha=0.3, label='Тренировочный') # тренировочный
plt.scatter(x_test[['bmi']],   y_test[['charges']],  alpha=0.3, label='Проверочный')   # проверочный

# Нанести график - линии, проведённые через точки - plt.plot
plt.plot( x_test[['bmi']], y_predicted_on_bmi,
          lw=2, color='red', alpha=0.3, marker='o',
          label="y_pred (X) = x0 * {:.3f} + {:.3f}".format(model_bmi.coef_[0][0], model_bmi.intercept_[0]))
plt.grid(True)  # Сетка. Доп параметры color='black', linewidth=0.7
plt.legend(loc="best")  # Положение легенды
plt.xlabel('BMI') # Подпись осей
plt.ylabel(target[0])   # Подпись осей
plt.title(u'Диаграмма рассеяния (точечная) для модели обученной на ЕДИНСТВЕННОМ признаке'); # Заголовок
```

Т.к. модель является линейной имеются некоторые расхождения между реальными и вычисленными значениями - "ошибки". Для оценки точности построим гистограмму распределения ошибок

```python
plt.figure(figsize=(6,4))
# Гистограмма и Плотность распределения - sns.distplot
# data = y_true - y_pred
sns.histplot( data = y_test[['charges']] - y_predicted_on_bmi,
              color="red",
              kde=True, # оценка плотности в виде кривой
              #stat="density",# density: общая площадь равна 1
)
plt.axvline(x=0, color='black', linestyle= 'dashed', linewidth=2,)  # Осевая линия
plt.ylabel(u'Кол-во')
plt.xlabel(u'Значение ошибки')
plt.grid(True, alpha=0.5) # Сетка
plt.title(u'Гистограмма ошибок');
```

Построим еще пару графиков для визуальной оценки качества модели:
- график рассеяния вычисленных значений Y_pred от проверочных значений Y_test;
- график квадратов ошибок от проверочных значний Y_test.

```python
def PLOT34(y_test, y_pred) -> None:
    '''
    Функция построения графиков
    :param y_test: - проверочные значения целевой переменной
    :param y_pred: - вычисленные значения целевой переменной
    '''
    plt.figure(figsize=(12,6))

    # == Диаграмма рассеяния вычисленных значений ==
    # создать первое полотно 121: 1-строк, 2-столбцов, 1-индекс текущего полотна в сетке
    plt.subplot(121)
    plt.scatter(y_test, y_pred,  alpha=0.1, color = "#17becf")
    plt.axline((0, 0), slope=1, color='black', linestyle='--', linewidth=3, alpha=0.7,)
    range_test, range_pred = np.max(y_test)- np.min(y_test) , np.max(y_pred)- np.min(y_pred)
    if range_test/range_pred <4 and range_test/range_pred > 0.25:
        plt.gca().set_aspect('equal')
        axmin, axmax = np.min([y_test, y_pred]), np.max([y_test, y_pred])
        plt.xlim([axmin, axmax]); plt.ylim([axmin, axmax]);
    plt.title('Диаграмма рассеяния вычисленных значений');
    plt.xlabel('Проверочное Y');    plt.ylabel('Вычисленное Y');
    plt.grid(True,  alpha=0.5)   # Сетка

    # == Диаграмма рассеяния ошибок ==
    # создать второе полотно 121: 1-строк, 2-столбцов, 2-индекс текущего полотна в сетке
    plt.subplot(122)
    plt.scatter(y_test, (y_test - y_pred)**2,  alpha=0.1, color = "#17becf")
    plt.title('Диаграмма рассеяния квадрата абсолютной ошибки')
    plt.xlabel('Проверочное Y');    plt.ylabel('Квадрат абсолютной ошибки');
    plt.grid(True,  alpha=0.5)  # Сетка
#--------------------------------------------------------------------------

PLOT34(
    y_test = y_test[['charges']],
    y_pred = y_predicted_on_bmi
)
```

С помощью функций __sklearn.metrics.*__ сравним реальные и вычисленные значениямя и получим важные оценки качества модели

```python
print ("Линейная одномерная регрессия. Признак 'YearBuilt'.")
print ("Формула линейной регресси y(X) = k0*x0 + ... kN*xN + b")
print(f"y = f(X) = ")
for i, kN in enumerate (model_bmi.coef_[0]):
    print(f" {' ' if i==0 else '+'}  {kN:12.3f} * x{i} ")
print(f"+ {model_bmi.intercept_[0]:.3f} ")
print("\n")
```

```python
'''
Вычисление и вывод метрик: MAE, RMSE, R2. На основе сравнения проверочных и вычисленных значений.
'''
mae  = metrics.mean_absolute_error(y_test, y_predicted_on_bmi)
mse  = metrics.mean_squared_error (y_test, y_predicted_on_bmi)
rmse = np.sqrt(mse)
r2   = metrics.r2_score           (y_test, y_predicted_on_bmi)
print ("MAE : {:>9,.3f} (средняя абсолютная ошибка)".format( mae ))
print ("MSE : {:>9,.6f} (среднеквадратичная ошибка)".format( mse ))
print ("RMSE: {:>9,.6f} (кв. корень из среднеквадратичной ошибки)".format( rmse ))
print ("R2  : {:>9,.3f} (коэфф. детерминации)".format      ( r2 ))
```
